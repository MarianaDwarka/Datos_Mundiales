```python
import mysql.connector
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style='whitegrid', palette = 'pastel')

# Configura la conexión a tu base de datos
config = {
    'user': 'root',       
    'password': '12345abcd', 
    'host': 'localhost',         
    'database': 'world' 
}

# Establece la conexión a la base de datos
conn = mysql.connector.connect(**config)
cursor = conn.cursor()

```

## DataFrame para city

```python
query = """
SELECT * FROM city;
"""

cursor.execute(query)
resultados1 = cursor.fetchall()

# Obtén los nombres de las columnas
column_names = [desc[0] for desc in cursor.description]

# Crea el DataFrame
city = pd.DataFrame(resultados1, columns=column_names)

# Muestra el DataFrame
city.head()
```

## DataFrame para country

```python
query = """
SELECT * FROM country
"""

cursor.execute(query)
resultados1 = cursor.fetchall()

# Obtén los nombres de las columnas
column_names = [desc[0] for desc in cursor.description]

# Crea el DataFrame
country = pd.DataFrame(resultados1, columns=column_names)

# Muestra el DataFrame
country.head()
```

## DataFrame para countrylanguage

```python
query = """
SELECT * FROM countrylanguage;
"""

cursor.execute(query)
resultados1 = cursor.fetchall()

# Obtén los nombres de las columnas
column_names = [desc[0] for desc in cursor.description]

# Crea el DataFrame
countrylanguage = pd.DataFrame(resultados1, columns=column_names)

# Muestra el DataFrame
countrylanguage.head()
```

# Visualizaciones

## Ejercicio 1: Consulta para mostrar el nombre y la población de todos los países del continente europeo

```python
# Filtrar el DataFrame para obtener solo los países de Europa
ejercicio_1 = country[country['Continent'] == 'Europe'][['Name', 'Population']]

# Mostrar el resultado
ejercicio_1
```

## Ejercicio 2: Consulta para mostrar los nombres y las áreas de superficie de los cinco países más grandes del mundo

```python
top5_countries = country[['Name', 'SurfaceArea']].sort_values(by='SurfaceArea', ascending=False).head(5)

top5_countries
```

## Ejercicio 3: Consulta para calcular la población total de todos los países de cada continente

```python
continent_population = country.groupby('Continent')['Population'].sum().reset_index()

continent_population
```

## Ejercicio 4: Consulta para mostrar el nombre de las ciudades y la población de todos los países de Europa

```python
# Realizar el join de los DataFrames 'city' y 'country' en la columna 'CountryCode' y 'Code' respectivamente
merged_df = pd.merge(city, country, left_on='CountryCode', right_on='Code').rename(columns={'Name_x': 'Name', 'Population_x': 'Population'})

# Filtrar el DataFrame para obtener solo las ciudades de Europa
european_cities = merged_df[merged_df['Continent'] == 'Europe']

# Seleccionar las columnas necesarias y ordenar por población en orden descendente
result_df = european_cities[['Name', 'Population']].rename(columns={'Name': 'CityName', 'Population': 'CityPopulation'}).sort_values(by='CityPopulation', ascending=False)

# Mostrar el resultado
result_df
```

## Pregunta 1: ¿Cuáles son los países con más ciudades en la base de datos?

```python
city_count = city.groupby('CountryCode').size().reset_index(name='CityCount')
city_count = city_count.merge(country[['Code', 'Name']], left_on='CountryCode', right_on='Code').drop(columns=['Code'])
top10_countries = city_count.sort_values(by='CityCount', ascending=False).head(10)

top10_countries
```

## Pregunta 2: ¿Cuál es la diversidad lingüística en términos de cantidad de idiomas hablados por país?

```python
# Realizar el join de los DataFrames 'country' y 'countrylanguage' en las columnas 'Code' y 'CountryCode' respectivamente
merged_df = pd.merge(country, countrylanguage, left_on='Code', right_on='CountryCode')

# Agrupar por el nombre del país y contar los idiomas
language_count_df = merged_df.groupby('Name').size().reset_index(name='LanguageCount')

# Ordenar por el conteo de idiomas en orden descendente
result_df = language_count_df.sort_values(by='LanguageCount', ascending=False).rename(columns={'Name': 'CountryName'})

# Mostrar el resultado
result_df
```

## Pregunta 3: ¿Cuáles son las 10 ciudades más pobladas del mundo?

```python
# Obtener las diez ciudades más pobladas
top10_cities = city[['Name', 'Population']].sort_values(by='Population', ascending=False).head(10)

top10_cities
```

## Pregunta 4: ¿Qué países tienen una expectativa de vida mayor a 80 años?

```python
high_life_expectancy = country[country['LifeExpectancy'] > 80][['Name', 'LifeExpectancy']]

high_life_expectancy
```

## Pregunta 5: ¿Cuáles son los continentes con la mayor cantidad de países?

```python
# Agrupar y contar el número de países por continente
continent_country_count = country.groupby('Continent').size().reset_index(name='CountryCount')

continent_country_count
```

## Pregunta 6: ¿Qué continentes tienen la mayor y menor densidad poblacional, y cuál es esa densidad?

```python
continent_density = country.groupby('Continent').apply(lambda x: round(x['Population'].sum() / x['SurfaceArea'].sum(),4)).reset_index(name='PopulationDensity')

continent_density
```

## Pregunta 7: ¿Cuál son los 3 idiomas más hablados en el mundo (considerando solo los datos disponibles)?

```python
language_speakers = country.merge(countrylanguage, left_on='Code', right_on='CountryCode')
language_speakers['TotalSpeakers'] = language_speakers['Population'] * language_speakers['Percentage'] / 100
language_speakers = language_speakers.groupby('Language')['TotalSpeakers'].sum().reset_index()
top3_languages = language_speakers.sort_values(by='TotalSpeakers', ascending=False).head(3)

top3_languages
```

## Pregunta 8: ¿Cuáles son los 3 países que tienen el mayor Producto Interno Bruto (PIB)?

```python
# Asegurarse de que la columna GNP es de tipo numérico
country['GNP'] = pd.to_numeric(country['GNP'], errors='coerce')

# Obtener los tres países con el mayor GNP
top3_gnp = country[['Name', 'GNP']].sort_values(by='GNP', ascending=False).head(3)

top3_gnp
```

## Pregunta 9: ¿Cuál es la proporción de habitantes que viven en la capital de cada país respecto a su población total?

```python
# Realizar la consulta y obtener los datos
merged_data = country.merge(city, left_on='Capital', right_on='ID', suffixes=('_country', '_city'))
merged_data['CapitalPopulationPercentage'] = (merged_data['Population_city'] / merged_data['Population_country']) * 100

# Ordenar los datos por CapitalPopulationPercentage y seleccionar los 10 primeros
top10_data = merged_data.sort_values(by='CapitalPopulationPercentage', ascending=False)[['Name_country', 'Name_city','Population_country', 'Population_city', 'CapitalPopulationPercentage']]

top10_data
```

## Pregunta 10: ¿Cuáles son los países con la mayor superficie terrestre en cada continente?

```python
# Asegurarse de que la columna SurfaceArea es de tipo numérico
country['SurfaceArea'] = pd.to_numeric(country['SurfaceArea'], errors='coerce')

# Realizar la consulta y obtener los datos
largest_countries = country.loc[country.groupby('Continent')['SurfaceArea'].idxmax()].reset_index(drop=True)[['Continent', 'Name', 'SurfaceArea']]

largest_countries
```
